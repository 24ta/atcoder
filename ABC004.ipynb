```
import numpy as np
a = [input().split() for i in range(4)]
a = np.array(a)
a = np.rot90(np.rot90(a))
# *によりアンパッキングして行列をきれいに表示
for i in a:
    print(*i)
#[print(*i) for i in a]
```

```
N = int(input())
Narray = list(range(1,7))
if N > 30:
  N = int(N%30)
for i in range(N):
  ii = (i % 5)
  iii = (i % 5) +1
  Narray[ii], Narray[iii] = Narray[iii], Narray[ii]
print(*Narray, sep="")
```
