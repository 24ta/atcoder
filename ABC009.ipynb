```
## 2時間やってわからなかったC問題
import copy
N,K = map(int, input().split())
S = list(input())
a=[]
for i in range(K):
  s = list(''.join(sorted(S[i:N])))
  if S[i] != s[0]:
    for j in reversed(range(i,N)):
      if s[0] == S[j]:
        S[i],S[j] = S[j],S[i]
        a.append(j)
        break
a.sort()
for k in range(len(a)):
  if len(a) == 1 or k == len(a)-1:
    break
  if S[a[k]] > S[a[k+1]]:
    S[a[k]],S[a[k+1]] = S[a[k+1]],S[a[k]]
print(''.join(S))
```

```
# 誰かのCの回答
from collections import Counter

N, K = map(int, input().split(' '))
S = input()
R = sorted(S)
T = ''
diff = 0

for i in range(N):
    # 各文字の出現回数をカウント
    counter = Counter(S[:i+1]) - Counter(T)
    for r in R:
        # 確定済み文字のdiff + みている文字のdiff
        diff1 = diff + (r != S[i])
        # 確定していない文字のdiff
        diff2 = sum(counter.values()) - (counter[r] > 0)
        if diff1 + diff2 <= K:
            T += r
            R.remove(r)
            diff = diff1
            break
print(T)
```
