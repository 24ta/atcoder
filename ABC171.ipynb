```
#26進数
N = int(input())
a=[]
answer=[]
for i in range(20):
  if (N > 26) and (N%26)!=0:
    a.append(N%26)
    N = int(N/26)
  elif (N > 26) and (N%26)==0:
    a.append(26)
    N = int((N-1)/26)
  else:
    a.append(N)
    break
for i in range(len(a)):
  answer.append(chr(ord('a')+a[i]-1))
mojiretu = ' '
for x in reversed(answer):
    mojiretu += x
print(mojiretu)
```

```
# dict
N = int(input())
A = [0] * N
tempA = list(map(int,input().split()))
for i in range(N):
  A[tempA[i]-1] = A[tempA[i]-1]+1

print(A)
Q = int(input())
for i in range(Q):
  S=0
  B,C = map(int, input().split())
  A[C-1] = A[C-1]+A[B-1]
  A[B-1] = 0
  for j in range(N):
    S = S+((j+1)*A[j])
  print(S)


# for i in range(Q):
#   B,C = map(int, input().split())
#   A = [C if j == B else j for j in A]
#   S = sum(A)
#   print(S)
```
