```
#26進数
N = int(input())
a=[]
answer=[]
for i in range(20):
  if (N > 26) and (N%26)!=0:
    a.append(N%26)
    N = int(N/26)
  elif (N > 26) and (N%26)==0:
    a.append(26)
    N = int((N-1)/26)
  else:
    a.append(N)
    break
for i in range(len(a)):
  answer.append(chr(ord('a')+a[i]-1))
mojiretu = ' '
for x in reversed(answer):
    mojiretu += x
print(mojiretu)
```

```
# dict
N = int(input())
A = list(map(int,input().split()))
Q = int(input())
for i in range(Q):
  B,C = map(int, input().split())
  A = [C if j == B else j for j in A]
  S = sum(A)
  print(S)
```
