```
N = int(input())
s = {"{}".format(input()): 1}
for i in range(N-1):
  temp = input()
  if temp in s:
    s['{}'.format(temp)] += 1
    continue
  s['{}'.format(temp)] = 1
s = sorted(s.items(), key=lambda x:x[1], reverse=True)
print(s[0][0])
```

```
import math
import itertools
from decimal import Decimal, ROUND_HALF_UP, ROUND_HALF_EVEN
N = int(input())
n = math.factorial(N)
c=[]
HorT=[]
sumc=[]
for i in range(N):
  c.append(int(input()))
  HorT.append(True)
for pair in itertools.permutations(c):
  count=0
  for i in range(N):
    for j in range(i+1,N):
      if pair[j]%pair[i] == 0:
        HorT[j] = not bool(HorT[j])
  for k in range(N):
    if HorT[k] == True:
      count=count+1
    HorT[k] = True
  sumc.append(count)
print(Decimal(str(sum(sumc)/n)).quantize(Decimal('0.000000000001'), rounding=ROUND_HALF_UP))
```
