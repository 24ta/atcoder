```
M,D = map(int, input().split())
if M % D == 0:
  print("YES")
else:
  print("NO")
```

```
A,B,C = map(int, input().split())
if (A+B) == C and (A-B) != C:
  print("+")
elif (A-B) == C and (A+B) != C:
  print("-")
elif (A+B) == C and (A-B) == C:
  print("?")
else:
  print("!")
```
