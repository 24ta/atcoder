```
from decimal import Decimal, ROUND_HALF_UP, ROUND_HALF_EVEN
N = int(input())
NN = list(range(10000,N*10000+1,10000))
NN = list(map(lambda x: x / N,NN))
print(int(Decimal(sum(NN)).quantize(Decimal('1'), rounding=ROUND_HALF_UP)))
```

```
S = input()
T = input()
lose = 0
for i in range(len(S)):
  if S[i] == '@' and T[i] in 'atcoder':
    S = S.replace(S[i], T[i], 1)
  elif S[i] == '@' and T[i] == '@':
    S = S.replace(S[i], 'a', 1)
    T = T.replace(T[i], 'a',1)
  elif T[i] == '@' and S[i] in 'atcoder':
    T = T.replace(T[i],S[i],1)
  if S[i] != T[i]:
    print('You will lose')
    lose = 1
    break
  elif S[i] == T[i]:
     continue
if lose == 0:
  print('You can win')
```
