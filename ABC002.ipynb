```
# ヘロンの公式
from math import sqrt
from decimal import Decimal, ROUND_HALF_UP, ROUND_HALF_EVEN
T = list(map(int, input().split()))
a = sqrt((T[2]-T[4])**2+(T[3]-T[5])**2)
b = sqrt((T[0]-T[4])**2+(T[1]-T[5])**2)
c = sqrt((T[2]-T[0])**2+(T[3]-T[1])**2)
s = (a+b+c)/2
S = sqrt(s*(s-a)*(s-b)*(s-c))
print(float(Decimal(S).quantize(Decimal('0.1'), rounding=ROUND_HALF_UP)))
```
