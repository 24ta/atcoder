```
n = int(input())
count = 0
flowers = list(map(int,input().split()))
for i in range(n):
  flower = flowers[i]
  for j in range(2):
    if flower % 2 == 0:
      count = count + 1
      flower = flower-1
    if (flower+1) % 3 == 0:
      count = count + 1
      flower = flower-1
print(count)
```

```
from math import sqrt
txa,txb,tya,tyb,T,V = map(int,input().split())
n = int(input())
isUwaki = False
for i in range(n):
  homex,homey = map(int,input().split())
  distance = sqrt((homex-txa)**2+(homey-txb)**2)+sqrt((tya-homex)**2+(tyb-homey)**2)
  move_dis = T*V
  if distance <= move_dis:
    print('YES')
    isUwaki = True
    break
if isUwaki == False:
  print('NO')
```
