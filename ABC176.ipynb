```
N = input()
sum = 0
for i in range(len(N)):
  sum = sum + int(N[i])
if(sum%9==0):
  print("Yes")
else:
  print("No")
```

```
N = int(input())
A = list(map(int, input().split()))
count=0
for i in range(N):
  if(i==0):
    continue
  if(A[i-1]>A[i]):
    count = count + A[i-1]-A[i]
    A[i] = A[i-1]
print(count)
```
