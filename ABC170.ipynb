```
X,N = map(int,input().split())
n = list(map(int,input().split()))
xn = list(range(0,102))
answer = 0
for i in range(N):
  xn.remove(n[i])
mindis = 100
for i in range(len(xn)):
  dis = abs(X-xn[i])
  if dis < mindis:
    mindis = dis
    answer = xn[i]
print(answer)
```

```
# 2020 6/15
# D
N = int(input())
A = list(map(int, input().split()))
count = 0
ans_count = 0
for i in range(N):
  for j in range(N):
    if A[i] == A[j]:
      continue
    if A[i]%A[j]!=0:
      count=count+1
  if count == (N-1):
    ans_count = ans_count + 1
  count = 0
print(ans_count)
```

```
from math import sqrt
import copy
N = int(input())
A = list(map(int, input().split()))
A.sort()
roop_end = int(sqrt(A[N-1]))
print(roop_end)
for i in range(len(A)):
  B = copy.copy(A)
  for j in B[i+1:]:
    print('{} {}'.format(B[i],j))
    if (j%B[i]) == 0 and B[i]!=j:
      A.remove(j)
    else:
      A.remove(j)
      A.remove(B[i])
  if A[i] > roop_end:
    break
print(len(A))
```
