```
# C
# 動的計画法による解答
N = int(input())
NG1= int(input())
NG2 = int(input())
NG3 = int(input())
dp = [float('inf')] * (N+1)
dp[N] = 0
for i in reversed(range(0, N+1)):
  if i == NG1 or i == NG2 or i == NG3:
    continue
  for j in range(1,4):
    if i-j < 0:
      break
    dp[i-j] = min(dp[i]+1, dp[i-j])
if dp[0] <= 100:
  print('YES')
else:
  print('NO')
```
