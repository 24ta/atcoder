```
T = int(input())
N = int(input())
A = list(map(int, input().split()))
A.sort()
M = int(input())
B = list(map(int, input().split()))
B.sort()
soldout = 0
for i in range(M):
  if M > N:
    soldout = 1
    break
  if A[0]>B[0]:
    soldout = 1
    break
  if T <= abs(A[0]-B[0]) and N>M:
    for j in range(N-M):
      del A[0]
      if len(A) < len(B):
        soldout = 1
        break
      if T >= abs(A[0]-B[0]):
        break
    del A[0]
    del B[0]
  elif T >= abs(A[0]-B[0]):
    del A[0]
    del B[0]
  else:
    soldout = 1
    break
  if (i+1) == M:
    break
if soldout == 0:
  print("yes")
elif soldout == 1:
  print("no")
```
