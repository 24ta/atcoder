```
import math
import time
stime = time.time()
sum = 0
K = 200
for k in range(K):
  for j in range(K):
    for i in range(K):
      sum += math.gcd(math.gcd(k+1,j+1),i+1)
print(sum)
etime = time.time() - stime
print(etime)
```

```
from math import gcd
import time
stime = time.time()
sum = 0
K = 200
for k in range(1,K+1):
  for j in range(1,K+1):
    t = gcd(k,j)
    for i in range(1,K+1):
      sum += gcd(t,i)
print(sum)
etime = time.time() - stime
print(etime)
```
