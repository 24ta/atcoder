```
n = int(input())
a = []
b = []
d = {'0': 0}
for i in range(1,1000000+1):
  d['{}'.format(i)] = 0

for i in range(n):
  ab = list(map(int, input().split()))
  for j in range(ab[0],ab[1]+1):
    d['{}'.format(j)] = d['{}'.format(j)] + 1
d_sorted = sorted(d.items(), key=lambda x:x[1], reverse=True)
print(d_sorted[0][1])
```

```
# いもす法　←　累積和の応用
n = int(input())
d = [0] * (1000000+2)
for i in range(n):
  a,b = map(int, input().split())
  d[a] = d[a]+1
  d[b+1] = d[b+1]-1
for i in range(1,1000000+2):
  d[i] = d[i]+d[i-1]
print(sorted(d,reverse=True)[0])
```
