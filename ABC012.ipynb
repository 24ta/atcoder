```
N = int(input())
if N > 3600:
  h = int(N/3600)
else:
  h = 0
if (N-3600) > 60 :
  m = int((N-3600*h)/60)
elif N > 60 and (N-3600) < 60:
  m = int(N/60)
else:
  m = 0
if N == 0:
  s=0
else:
  s = int(N-3600*h-60*m)
h = str(h).zfill(2)
m = str(m).zfill(2)
s = str(s).zfill(2)
print('{}:{}:{}'.format(h,m,s))
```

```
N = int(input())
sumKuku = 0
for i in range(1,10):
  for j in range(1,10):
    sumKuku = sumKuku + (i*j)
M = sumKuku - N
for i in range(1,10):
  for j in range(1,10):
    if (i*j) == M:
      print('{} x {}'.format(i,j))
```
